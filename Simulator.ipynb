```python
# Hadamard Gate - Simulator

This notebook demonstrates how a Hadamard gate creates superposition using a quantum simulator.

from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
import numpy as np

# Create circuit
qc = QuantumCircuit(1)

# Apply Hadamard gate
qc.h(0)

# Get statevector
state = Statevector.from_instruction(qc)

# Probabilities
probs = state.probabilities_dict()

print(probs)
```

```python
## Result

The simulator shows equal probability for both states:

- P(0) = 0.5  
- P(1) = 0.5  

This represents perfect superposition.
```
